Question 1:
Compute the density of the data using a histogram with 5 bins

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("/content/heart_attack_dataset.csv")
data.head()
```

```python
Age=data['Age']
Age.head()
```

Compute the density of the data using a histogram with 5 bins

```python
num_bins = 5
hist, bin_edges = np.histogram(Age, bins=num_bins, density=True)

plt.figure(figsize=(8, 6))
plt.hist(Age,bins=bin_edges, density=True, edgecolor='black', alpha=0.6, color='yellow')
plt.title('Histogram of Age with 5 bins')
plt.xlabel('Age Values')
plt.ylabel('Density')
plt.grid(True)
plt.show()
print("Histogram values (density):", hist)
print("Bin edges:", bin_edges)
```

Question 2:

Find Mean,Median,Mode,Variance,Standard deviatio,Range,Interquartile Range(IQR),Skewness,kurtosis of blood pressure(mmHg)
and cholestral(mg/dl)

```python
x = data["Blood Pressure (mmHg)"]
y = data["Cholesterol (mg/dL)"]
```

```python
from scipy import stats
import statistics as st

def calculate_statistics(data):
    # Convert to a pandas Series
    data_series = pd.Series(data)

    # Mean
    mean =  np.mean(data)

    # Median
    media = np.median(data)

    # Mode
    mode = st.mode(data)

    # Variance
    variance = np.var(data, ddof=1)  # sample varience

    # Standard Deviation
    std_deviation = np.std(data, ddof=1)  # sample standard deviation

    # Range
    data_range = np.max(data) - np.min(data)

    # Interquartile Range (IQR)
    iqr = stats.iqr(data)

    # Skewness
    skewness = stats.skew(data)

    # Kurtosis
    kurtosis = stats.kurtosis(data)

    return {
        "Mean": mean,
        "Median": media,
        "Mode": mode,
        "Variance": variance,
        "Standard Deviation": std_deviation,
        "Range": data_range,
        "Interquatile Range(IQR)": iqr,
        "Skewness": skewness,
        "Kurtosis": kurtosis
    }

 # Calculate statistics for x and y
stats_x = calculate_statistics(x)
stats_y = calculate_statistics(y)

 # Print the results
print("Statistics for Dependent Variables (x):")
for key, value in stats_x.items():
    print(f"{key}: {value}")

print("\nStatistics for Independent Variables (y):")
for key, value in stats_y.items():
    print(f"{key}: {value}")
```
